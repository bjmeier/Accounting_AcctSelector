```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
#SOURCE_FILE_FOLDER = 'Q:\OdooMigration\Accounting\costCenterInputs\\'
#SOURCE_FILE_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\assetInputs\\'
#SOURCE_CC_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\costCenterInputs\\'
SOURCE_FILE_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\'
import datetime
now = datetime.datetime.now()
print(f"Report generated at {now.strftime('%H:%M:%S on %A, %B the %dth, %Y')}.")
```

```python
fname = 'ChartOfAccountStructure.xlsx'
acctSheetNames = ['Assets', 
              'Liabilities', 
              'Equity', 
              'Revenue', 
              'Product Expense', 'CC Expense', 'COGS Expense', 'Building Expense', 'SGA expense'
             ]
```

```python
dfs = []
clist = ['Code','GLAcctNo','OdooAcctNo','Parent']
for i in range(len(acctSheetNames)):
    df = pd.read_excel(SOURCE_FILE_FOLDER+fname, sheet_name=acctSheetNames[i], index_col=0)
    for c in clist:
        try:
            df[c]=df[c].astype('Int64')
        except:
            print(f'could not convert {acctSheetNames[i]} col: {c}')
        if i==0:
            pass
        else:
            df['GLAcctNo'] = df['GLAcctNo'].fillna("")
        df['Notes'] = df['Notes'].fillna("")
    dfs.append(df)
```
