```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
#SOURCE_FILE_FOLDER = 'Q:\OdooMigration\Accounting\costCenterInputs\\'
SOURCE_FILE_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\assetInputs\\'
SOURCE_CC_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\costCenterInputs\\'
SOURCE_TOP_FOLDER = 'C:\\Users\\bmeie\\Documents\\Accounting\\'
import datetime
now = datetime.datetime.now()
print(f"Report generated at {now.strftime('%H:%M:%S on %A, %B the %dth, %Y')}.")
```

```python
def saveAsCsv(df, fname, folder=SOURCE_FILE_FOLDER):
    # Saves dataframe as a csv file
    # example call: saveAsCsv(buildDf, 'build.csv')
    df.to_csv(folder+f'{fname}',index=None)
```

```python
def formatNums(dfInp):
    df=dfInp.copy()
    lst = list(df.columns)
    for c in lst:
        try:
            df[c] = df[c].astype(int).apply(lambda x: f'{x:,}')
        except:
            pass
    return df
```

```python
codeDescDf     = pd.read_csv(SOURCE_CC_FOLDER +'codeDesc.csv')  
assetDf        = pd.read_csv(SOURCE_FILE_FOLDER +'Assets.csv') 
classCodeDescExpCodesDf        = pd.read_csv(SOURCE_FILE_FOLDER +'classCodeDescExpCodes.csv') 
currentClassCodeFutureClassCodeDf        = pd.read_csv(SOURCE_FILE_FOLDER +'currentClassCodeFutureClassCode.csv') 
currentCodeDescDefaultDf        = pd.read_csv(SOURCE_FILE_FOLDER +'currentCodeDescDefault.csv') 
assetIDclassCodeDf = pd.read_csv(SOURCE_FILE_FOLDER +'assetIDclassCode.csv') 
eCodeExpAcctDf = pd.read_csv(SOURCE_FILE_FOLDER +'eCodeExpAcct.csv')
```

## Setting AssetID-driven Chart of Accounts

```python
df = classCodeDescExpCodesDf.copy()
dfExp = codeDescDf.copy()
dfExp.set_index('Code',inplace=True)
nameList = []
cCodeList = []
eCodeList = []
ceCodeList = []
parentList = []
acctNoList = []
parentAcctList = []
for i in range(len(df)):
    cCode = df.loc[i,'classCode']
    cDesc = df.loc[i,'classCodeDesc']
    eCodes = df.loc[i,'expCodes'].split()
    acctNo = 15000+cCode*100
    parentAcctNo = str(acctNo)
    nameList.append(cDesc)
    cCodeList.append(cCode)
    eCodeList.append("")
    ceCode = cCode*1000
    ceCodeList.append(ceCode)
    parent=str(int(ceCode))
    parentAcctList.append("")
    parentList.append("")
    acctNoList.append(acctNo)
    for ec in eCodes:
        acctNo += 1
        eDesc = dfExp.loc[int(ec),'Desc']
        name = f'{cDesc} - {eDesc}'
        nameList.append(name)
        cCodeList.append(cCode)
        eCodeList.append(ec)
        ceCodeList.append(cCode*1000+int(ec))
        parentList.append(parent)
        acctNoList.append(acctNo)
        parentAcctList.append(parentAcctNo)

dfAssetIDAccounts = pd.DataFrame()
dfAssetIDAccounts['Desc'] = nameList
dfAssetIDAccounts['Type']='Non-current Assets'
dfAssetIDAccounts['Code'] = ceCodeList
dfAssetIDAccounts['odooAcctNo'] = acctNoList
dfAssetIDAccounts['Parent']=parentAcctList
dfAssetIDAccounts['cCode'] = cCodeList
dfAssetIDAccounts['eCode'] = eCodeList
dfAssetIDAccounts['parentCode'] = parentList
dfAssetIDAccounts.sort_values('odooAcctNo', inplace=True)
dfAssetIDAccounts.reset_index(inplace=True, drop=True)
#saveAsCsv(dfAssetIDAccounts, 'dfAssetIDaccounts.csv')
dfAssetIDAccounts
```

```python
adf = assetIDclassCodeDf.merge(currentClassCodeFutureClassCodeDf,left_on='currentClassCode',
                              right_on='currentClassCode',how='outer')
adf = adf.merge(currentCodeDescDefaultDf,
               left_on='currentClassCode',
               right_on='currentClassCode',
               how='left')

adf = adf[['AssetID','classCode','defaultCode','defaultExp']]
adf = assetDf.merge(adf, left_on='AssetID', right_on='AssetID', how='outer')
adf.rename(columns={'Code':'eCode', 'classCode': 'cCode'}, inplace=True)
adf['ceCode']=adf['cCode']*1000+adf['eCode']
adf.loc[(adf.ceCode == 5900),'ceCode']=5906
adf.loc[(adf.ceCode == 5370),'ceCode']=5906
adf.loc[(adf.ceCode == 5906),'eCode']=906
cCodeAcctDf = dfAssetIDAccounts.copy()
cCodeAcctDf = cCodeAcctDf[['Code', 'odooAcctNo']]
cCodeAcctDf.rename(columns={'Code':'ceCode'},inplace=True)
adf=adf.merge(cCodeAcctDf, left_on='ceCode', right_on='ceCode',how='left')
adf.rename(columns={'odooAcctNo':'odooAssetAcctNo'}, inplace=True)
adf.loc[(adf.DeprYTD==0),'eCode']=adf.loc[(adf.DeprYTD==0),'defaultCode']
adf = adf.merge(eCodeExpAcctDf, 
         left_on ='eCode',
         right_on='AssetCode',
         how='left')
adf = adf[['AssetID', 'Desc', 'DeprYTD','cCode', 'eCode','ceCode','odooAssetAcctNo','OdooAcctNo',
           'defaultCode','defaultExp']]
adf.rename(columns={'OdooAcctNo':'odooExpAcctNo'},inplace=True)
adf.loc[(adf.DeprYTD==0),'odooExpAcctNo']=adf.loc[(adf.DeprYTD==0),'defaultExp']
adf['odooExpAcctNo'] = adf['odooExpAcctNo'].astype(int)
adf = adf[['AssetID', 'Desc', 'DeprYTD','cCode', 'eCode','ceCode','odooAssetAcctNo','odooExpAcctNo']]

adf
```

```python
saveAsCsv(adf,'AssetAccts.csv')
```

```python
cCodeDf = currentCodeDescDefaultDf.copy()
cCodeDf = cCodeDf[['currentClassCode','CurrentClassDesc']]
cCodeDf.rename(columns={'currentClassCode': 'cCode', 'CurrentClassDesc': 'cDesc'},inplace=True)

eCodeDf = codeDescDf.copy()
eCodeDf.rename(columns={'Code': 'eCode', 'Desc': 'eDesc'},inplace=True)

dfOdoo = pd.read_pickle(SOURCE_TOP_FOLDER+'dfOdoo.pkl')
acctDescDf = dfOdoo.copy()
acctDescDf.reset_index(inplace=True)
acctDescDf=acctDescDf[['OdooAcctNo','Desc']]
acctDescDf.rename(columns={'Desc':'acctDesc'},inplace=True)

adfView = adf.merge(cCodeDf, 
                   left_on='cCode',
                   right_on='cCode',
                   how='left').merge(eCodeDf, left_on='eCode', right_on='eCode', how='left')

adfView = adfView[['AssetID', 'Desc', 'DeprYTD','cCode','cDesc','eCode','eDesc','odooAssetAcctNo','odooExpAcctNo']]
newDf = adfView.merge(dfAssetIDAccounts[['odooAcctNo','Desc']],left_on='odooAssetAcctNo',right_on='odooAcctNo',how='left')
newDf.rename(columns={'Desc_x':'Desc','Desc_y':'assetAcctDesc'},inplace=True)
newDf = newDf.set_index('odooExpAcctNo').join(acctDescDf.set_index('OdooAcctNo')).reset_index()
newDf.rename(columns={'index':'odooExpAcctNo','acctDesc':'expAcctDesc'},inplace=True)
newDf = newDf[['AssetID','Desc','DeprYTD','cCode','cDesc','eCode','eDesc','odooAssetAcctNo','assetAcctDesc','odooExpAcctNo','expAcctDesc']]
newDf.sort_values(by=['cCode','eCode'],inplace=True)
newDf
```

```python
saveAsCsv(newDf,'AssetsWithDesc.csv')
```
